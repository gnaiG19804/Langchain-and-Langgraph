```python
from dotenv import load_dotenv
load_dotenv()

import os
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain_core.messages import AIMessage, HumanMessage
from pprint import pprint

messages= [AIMessage(content=f"Please tell me how can I help you",name="LLMModel")]
messages.append(HumanMessage(content="tôi muốn học về lập trình",name="GIANG"))
messages.append(AIMessage(content=f"Bạn muốn học về ngôn ngữ lập trình nào",name="LLMModel"))
messages.append(HumanMessage(content="Tôi muốn học về Python",name="GIANG"))

for message in messages:
    message.pretty_print()
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model="llama-3.3-70b-versatile")
result= llm.invoke(messages)
```
